```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/kilinandrea/FDA_recall_project/main/FDA_recall_clean.csv')
```

```python
df.head()
```

```python
# Remove NaN values from the 'termination_speed' column
speed_data = df['termination_speed'].dropna()

# Plot the histogram
plt.hist(speed_data, bins=20, color='#00886F', edgecolor='black')

# Add labels and title
plt.xlabel('Termination Speed')
plt.ylabel('Frequency')
plt.title('Histogram of Termination Speed')

# Show the plot
plt.show()
```

```python

# Calculate the number of occurrences for each classification
classification_counts = df['classification'].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(classification_counts, labels=classification_counts.index, autopct='%1.1f%%', startangle=90, colors=['#00886F','#429DFF','#FFF52D', '#FF5900'])
plt.title('Distribution of FDA Recall Classifications')

# Show the plot
plt.show()
```

```python

# Calculate the number of occurrences for each status
status_counts = df['status'].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', startangle=90, colors=['#00886F','#429DFF','#FFF52D', '#FF5900'])
plt.title('Distribution of FDA Recall Statuses')

# Show the plot
plt.show()
```

```python
# Calculate the number of occurrences for each recall category
recall_category_counts = df['recall_category'].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(recall_category_counts, labels=recall_category_counts.index, autopct='%1.1f%%', startangle=90, colors=['#00886F','#429DFF','#FFF52D', '#FF5900', '#FFDECC', '#99CFC5', '#BABABB', '#7683A2', '#7C6E12'])
plt.title('Distribution of FDA Recall Categories')

# Show the plot
plt.show()

```

```python
# Calculate the number of occurrences for each notification type
notification_type_counts = df['notification_type'].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(notification_type_counts, labels=notification_type_counts.index, autopct='%1.1f%%', startangle=90, colors=['#00886F','#429DFF','#FFF52D', '#FF5900', '#FFDECC', '#99CFC5', '#BABABB', '#7683A2', '#7C6E12'])
plt.title('Distribution of FDA Recall Notification Types')

# Show the plot
plt.show()
```

```python
# Calculate the number of occurrences for each manufacturer
manufacturer_counts = df['manufacturer_name'].value_counts().head(10)

# Plot the bar plot
plt.figure(figsize=(10, 6))
manufacturer_counts.plot(kind='bar', color='#00886F')
plt.title('Top 10 Manufacturers by Frequency')
plt.xlabel('Manufacturer')
plt.ylabel('Frequency')

# Show the plot
plt.show()

```
